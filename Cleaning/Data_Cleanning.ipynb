 ### e) Return the number of columns for each quartiles for the % of null values


```python
def pct_null_values(df):
    return (df.isnull().sum())/ len(df)*100

col_list = food_facts.columns.tolist()
quart_1 = [x for x in col_list if pct_null_values(food_facts[x]) < 25]
quart_2 = [x for x in col_list if pct_null_values(food_facts[x]) >= 25 and pct_null_values(food_facts[x]) < 50]
quart_3 = [x for x in col_list if pct_null_values(food_facts[x]) >= 50 and pct_null_values(food_facts[x]) < 75]
quart_4 = [x for x in col_list if pct_null_values(food_facts[x]) >= 75]

quart_4
```

 ### f) percentage of null by column
 ### Function that take a df and return the % of non null values:

```python
# adding a column with the filling percentage
def filling_factor(df):
    missing_df = df.isnull().sum(axis=0).reset_index()
    missing_df.columns = ['column_name', 'missing_count']
    missing_df['filling_factor'] = (df.shape[0]-missing_df['missing_count'])/df.shape[0]*100
    missing_df = missing_df.sort_values('filling_factor').reset_index(drop = True)
    return missing_df
#____________________________________
missing_df = filling_factor(food_facts)
missing_df[missing_df['filling_factor'] == 0]
```

 ### f) chart with the filling factor

```python
col_filling = filling_factor(food_facts)
y_axis  = list(col_filling['filling_factor'])
x_axis  = [i for i in range(len(col_filling))]
x_label = list(col_filling['column_name'])
fig = plt.figure(figsize=(10, 22))

fig.patch.set_facecolor('white') # couleur de fond de figure qui contient le graphe
plt.yticks(x_axis, x_label)

plt.title('Filling factor (%)', fontsize = 15)
plt.barh(x_axis, y_axis)

plt.show()
```

 ### g) filling percentage by row and printing row with missing values

```python
def filling_factor_row(df):
    missing_df = df.isnull().sum(axis = 1).reset_index()
    missing_df.columns = ['column_name', 'missing_count_row']
    missing_df['filling_factor_row'] = (df.shape[1]-missing_df['missing_count_row'])/df.shape[1]*100
    missing_df = missing_df.sort_values('filling_factor_row').reset_index(drop = True)
    return missing_df
missing_row_df = filling_factor_row(food_facts)
missing_row_df[missing_row_df['filling_factor_row'] > 0]
```

 ## II Cleaning dataset:

 ### a) selecting rows for France only:

```python
food_facts_fr = food_facts[(food_facts['countries']=='France') |(food_facts['countries'] =='en:FR')]
# changing
food_facts_fr.shape
```

 ### b) Removing columns with less than 60% of null values

```python
col_less_60 = [x for x in col_list if pct_null_values(food_facts_fr[x]) >60]
print(len(col_less_60))

print(col_less_60)
print(len(col_less_60))

# Removing the following columns:
col_to_remove = ['code','url','last_modified_datetime','creator', 'created_t', 'created_datetime', 'last_modified_t', 'brands_tags', 'countries', 'countries_tags', 'additives_tags', 'states', 'states_tags']
print(len(col_to_remove))
print(food_facts_fr.shape)
food_facts_fr.drop(labels=col_to_remove, axis=1, inplace=True)
print(food_facts_fr.shape)
```

 ### c) Removing col_to_remove from the df:

```python
duplicate_cols = list(set(col_to_remove).intersection(col_less_60))
print(len(col_less_60))
for col in duplicate_cols:
    col_less_60.remove(col)
print(len(col_less_60))
```

 ### d) Removing col_less_60 from the df

```python
food_facts_fr = food_facts_fr.drop(labels=col_less_60, axis=1)
print(food_facts_fr.shape)
```

 ### c) chart of the new dataframe with the filling percentage:

```python
col_filling = filling_factor(food_facts_fr)
y_axis  = list(col_filling['filling_factor'])
x_axis  = [i for i in range(len(col_filling))]
x_label = list(col_filling['column_name'])
fig = plt.figure(figsize=(10, 22))

fig.patch.set_facecolor('white') # couleur de fond de figure qui contient le graphe
plt.yticks(x_axis, x_label)

plt.title('Filling factor (%)', fontsize = 15)
plt.barh(x_axis, y_axis)
plt.savefig("High resoltion.png",dpi=300)
plt.show()
```

 ### f) function to drop rows with null value

```python
def drop_row(df,col):
# drop row with null product_name
    df_toDrop=df[df[col].isnull()]# select rows with null product_name
    print('Rows to drop : ', df_toDrop.shape[0])
    df_droped=df.drop(df_toDrop.index, axis=0) #dropp rows
    return df_droped
#df_cleaned_row=drop_row(df_drop,'countries')
df_cleaned_row=drop_row(food_facts_fr,'product_name')
df_cleaned_row.shape
```

 ### g) drop duplicate rows (same brand and product name)

```python
df_lower=food_facts_fr.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
df_duplicates= df_lower[df_lower.duplicated(subset = ['product_name', 'brands'])]
food_facts_fr=df_lower.drop(df_duplicates.index, axis=0)
print(food_facts_fr.shape)
```
